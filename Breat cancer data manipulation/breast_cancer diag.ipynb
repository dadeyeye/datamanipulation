Question 1: Creating artificial missingness by randomly removing 25% of values

```python
temp = [(row, col) for row in range(df1.shape[0]) for col in range (df1.shape[1])]
for row, col in random.sample(temp, int(round(.25*len(temp)))):
    df1.iat[row,col] = np.nan
df1
```
